```python
Nigeria=Malaria_Africa[Malaria_Africa['Country Name']=='Nigeria']
Nigeria.head()
```

```python
import plotly.express as px
```

```python
#line graph for malaria cases in Nigeria between 2007 to 2017
fig = px.line(Nigeria, x="Year", y="Malaria cases reported", title='Malaria cases reported in Nigeria from 2007-2017')
fig.show()
```

```python
fig = px.scatter(Nigeria, x="Year", y="People using safely managed drinking water services, rural (% of rural population)", title='people using safely dug water in nigeria')
fig.show()
```

```python
# To find malaria cases recorded per country
malaria_cases_country= Malaria_Africa[['Country Name','Malaria cases reported']].groupby('Country Name').sum().sort_values(by='Malaria cases reported')
malaria_cases_country
```

```python
#since the information is to bulky for the malaria cases in the country, we would pcik the last 7
malaria_cases_country.tail(7)
```

```python
e= malaria_cases_country.tail(7)
e["Country Name"] = ['Ghana','Nigeria','Burundi','Uganda','Burkina Faso','Mozambique','Congo, Dem. Rep']
e
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x="Malaria cases reported", y="Country Name", data=e)
```
